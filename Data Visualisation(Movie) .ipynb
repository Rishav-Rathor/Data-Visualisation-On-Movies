# Data Visualisation challenge
### On Movie dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read a csv file
df = pd.read_csv("./movie_metadata.csv")
df.head(n=10)
```

```python
df.columns
```

```python
titles = list(df.get("movie_title"))
print(type(titles))
print(titles[:5])
```

```python
freq_titles = {}

for title in titles:
    length = len(title)

    if freq_titles.get(length) is None:
        freq_titles[length] = 1
    else:
        freq_titles[length] += 1
```

```python
print(freq_titles)
```

```python
x = np.array(list(freq_titles.keys()))
y = np.array(list(freq_titles.values()))
print(x,y)
```

```python
plt.scatter(x,y)
plt.xlabel("Length of Movie Title")
plt.ylabel("Freq Count")
plt.title("Movie data visualisation")
plt.show()
```
